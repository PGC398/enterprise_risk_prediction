```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv('selected/train_nofill.csv')
```

```python
train.describe()
```

```python
b=[]
for i in range(len(train)):
    try:
        b.append(np.isnan(train.iloc[i]).value_counts()[True])
    except:
        b.append(0)
```

```python
xx=[x for x in range(len(train))]
```

```python
import matplotlib.pyplot as plt
plt.plot(xx, sorted(b),'ro')
plt.title("train set") 
plt.xlabel("Order Number")
plt.ylabel("Number of Missing Attributes")
```

```python
plt.show()
```

```python
test=pd.read_csv('selected/test_nofill.csv')
```

```python
test.describe()
```

```python
c=[]
for i in range(len(test)):
    try:
        c.append(np.isnan(test.iloc[i]).value_counts()[True])
    except:
        c.append(0)
```

```python
xxx=[x for x in range(len(test))]
```

```python
plt.plot(xxx, sorted(c),'ro')
plt.title("test set") 
plt.xlabel("Order Number")
plt.ylabel("Number of Missing Attributes")
plt.show()
```
